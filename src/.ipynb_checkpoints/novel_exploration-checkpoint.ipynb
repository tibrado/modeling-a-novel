```python
# Imports 
import pandas as pd
import numpy as np
import shutil
import glob
import os
```

```python
# Get glossary 
column_names = ['title', 'chapters', 'stars', 'tags', 'followers', 'pages', 'views', 'date']
glossary =  pd.read_csv('../data/index.csv' ,delimiter=';', names = column_names)
```

```python
glossary
```

```python
# get all unique genre s
genres = dict()
def unique_genres(x):
    for g in x.split(' '):
        g = g.lower()
        if g not in genres:
            genres[g] = 0
        genres[g] += 1
        
glossary.tags.apply(lambda x: unique_genres(x))

genres
```

```python
# Remove pointless keys 
del genres['fantasy']
del genres['/']
del genres['lead']
del genres['of']
del genres['original']
```

```python
# round start up 
glossary['stars'] = glossary.stars.apply(lambda x: round(x, 1))
```

## Select Books of Intrest 

```python
# Get unique genere for model 
dystopia = glossary[glossary.tags.apply(lambda x: True if 'dystopia' in x.lower() else False)]
dystopia.describe().T[['mean', 'min', 'max', 'count']]
```

```python
# Make training data set for gpt2 
# simply copy books to other folder 
def make_training_book(indices, name):
    # make directroy 
    path = f'../data/train/{name}'
    os.mkdir(path)
    
    for book in indices:
        shutil.copy2(f'../data/books/book_{book + 1}', path)
```

```python
make_training_book(dystopia.index, 'dystopia')
make_training_book(litrpg.index, 'litrpg')
make_training_book(martial.index, 'martial')
make_training_book(dungeon.index, 'dungeon')
make_training_book(time.index, 'time')
```
